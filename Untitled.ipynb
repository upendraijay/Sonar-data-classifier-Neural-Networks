# Introduction
This dataset consists of sonar data. The aim of this data is do distinguish between rocks and metal structures such as sea mines on the seafloor. The file 111 patterns obtained by bouncing sonar signals off a metal cylinder at various angles and under various conditions and 97 patterns obtained from rocks under similar conditions.

Each pattern is a set of 60 numbers in the range 0.0 to 1.0. The label associated with each record contains the letter "R" if the object is a rock and "M" if it is a mine (metal cylinder).

Data can be downloded from: https://archive.ics.uci.edu/ml/machine-learning-databases/undocumented/connectionist-bench/sonar/sonar.all-data

## Loading the data
To load the data and format it nicely, we will use two very useful packages called Pandas and Numpy.

```python
# Importing pandas and numpy
import pandas as pd
import numpy as np
from keras.utils import to_categorical
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/undocumented/connectionist-bench/sonar/sonar.all-data', header=None, prefix='X')
df[:5]
```

## Splitting the data into features and targets (labels)
Now, we'll split the data into features (X) and targets (y).


```python
#create a dataframe with all training data except the target column
X = df.drop(columns=['X60'])
#check that the target variable has been removed
X.head()
```

#### As the target is a string value so we convert it into string using label encoder

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
Y=le.fit_transform(df.X60)
print(Y)
```

## Splitting the data into Train and Test 

```python
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(X, Y, test_size = 0.2, random_state = 1)
```

## Feature Scaling

```python

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xTrain = sc.fit_transform(xTrain)
xTest = sc.transform(xTest)
```

# Defining the Baseline model architecture
Here's where we use Keras to build our neural network.

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(60, input_dim=60, activation='relu'))
model.add(Dense(30,  activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## Train the Model

```python
model.fit(xTrain, yTrain, epochs=200, batch_size=16)
```

## Evaluate the Model to get the loss and accuracy

```python
score = model.evaluate(xTest, yTest, verbose=0)
print(score)
```

## Conclusion
As we can see that training loss is very low as compared to test loss. So our mode is overfitted for training data. So we will try to optimize it using the early Stopping technique.


```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
model.add(Dense(60, input_dim=60, activation='relu'))
model.add(Dense(30,  activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## Set early stopping monitor so the model stops training when it starts overfitting

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint
filepath='weights.best.hdf5'
keras_callbacks=[
    EarlyStopping(monitor='val_loss',patience=10),
    ModelCheckpoint(filepath,monitor='val_loss',save_best_only=True)
]
```

## Train the Model with early stopping

```python
model.fit(xTrain, yTrain, epochs=200, validation_split=0.2, batch_size=16, callbacks=keras_callbacks)
```

## Load and evaluate the best model saved by checkpoint callback

```python
from keras.models import load_model
saved_model = load_model('weights.best.hdf5')
score = saved_model.evaluate(xTest, yTest, verbose=0)
#Returns the loss value & metrics values for the model in test mode.
print(score)
```

## Conclusion
we can see that after adding early stopping testing loss is reduced from 1.5987673827580042 to 0.42842938183319
Now to further reduce this we can make the use of dropout technique.

```python
from keras.layers import Dropout
from keras.constraints import maxnorm
model = Sequential()
model.add(Dropout(0.2, input_shape=(60,)))
model.add(Dense(60, activation='relu',kernel_initializer='normal', kernel_constraint=maxnorm(3)))
model.add(Dropout(0.2))
model.add(Dense(30, activation='relu',kernel_initializer='normal', kernel_constraint=maxnorm(3)))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid',kernel_initializer='normal'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## Set early stopping monitor so the model stops training when it starts overfitting
- this time we use large patience value are we are using dropout so validation loss fluctuate 

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint
filepath='weights.best.hdf5'
keras_callbacks=[
    EarlyStopping(monitor='val_loss',patience=25),
    ModelCheckpoint(filepath,monitor='val_loss',save_best_only=True)
]
```

## Train the Model with early stopping and dropOut

```python
model.fit(xTrain, yTrain, epochs=300, validation_split=0.2, batch_size=16, callbacks=keras_callbacks)
```

## Load and evaluate the best model saved by checkpoint callback

```python
from keras.models import load_model
saved_model = load_model('weights.best.hdf5')
score = saved_model.evaluate(xTest, yTest, verbose=0)
print(score)
```

## Conclusion
we can see that after adding dropout stopping testing loss is reduced from 0.42481859241213116 to 0.4127456545829773. this is very small progress because the size of the data is small. so we use dropout when the size of data is large.
